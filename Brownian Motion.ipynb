```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
T = 10.0    # Total time
N = 100000   # Number of steps
dt = T / N # Time step size

# Generate increments
dW = np.random.normal(0, np.sqrt(dt), N)

# Generate the Brownian motion path
W = np.cumsum(dW)

# Add the initial value (W(0) = 0)
W = np.insert(W, 0, 0)

# Time vector
t = np.linspace(0, T, N+1)

# Compute the variance at each time step
variance = np.zeros(N+1)
for i in range(1, N+1):
    variance[i] = np.var(W[:i+1])

# Plot the Brownian motion path
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, W, label='Brownian Motion')
plt.xlabel('Time')
plt.ylabel('W(t)')
plt.title('Sample Path of Brownian Motion')
plt.legend()
plt.grid(True)

# Plot the computed variance
plt.subplot(2, 1, 2)
plt.plot(t, variance, label='Computed Variance', color='orange')
plt.xlabel('Time')
plt.ylabel('Variance')
plt.title('Variance of Brownian Motion')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
